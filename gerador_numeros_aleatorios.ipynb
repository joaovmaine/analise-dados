````python
import random
import ipywidgets as widgets
from IPython.display import display, clear_output

def gerar_numero_aleatorio(inicio, fim):
    """
    Gera um n√∫mero inteiro aleat√≥rio entre 'inicio' e 'fim'
    """
    return random.randint(inicio, fim)

# Criando widgets para definir o intervalo
inicio_slider = widgets.IntSlider(min=1, max=100, description='In√≠cio:')
fim_slider = widgets.IntSlider(min=1, max=100, value=100, description='Fim:')
button = widgets.Button(description='Gerar n√∫mero', button_style='info')

# Estilizando o bot√£o
button.style.button_color = 'lightblue'

# Output para exibir o resultado
output = widgets.Output()

# Fun√ß√£o para atualizar e exibir o resultado
def on_button_click(b):
    inicio = inicio_slider.value
    fim = fim_slider.value
    numero_aleatorio = gerar_numero_aleatorio(inicio, fim)

    with output:
        output.clear_output()
        print(f"N√∫mero aleat√≥rio entre {inicio} e {fim}: {numero_aleatorio}")

# Associando a fun√ß√£o ao bot√£o
button.on_click(on_button_click)

# Criando um layout mais organizado usando o Vbox
vbox = widgets.VBox([inicio_slider, fim_slider, button, output])

# Estilizando o layout
vbox.layout.align_items = 'center'
vbox.layout.margin = '50px 100px'

# Exibindo o layout
display(vbox)
````
